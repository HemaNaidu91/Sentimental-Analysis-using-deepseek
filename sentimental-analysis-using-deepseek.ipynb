```python
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
def predict_sentiment(text):
    from transformers import pipeline
    classifier = pipeline("sentiment-analysis")
    result = classifier(text)

    label_map = {
        "NEGATIVE": "negative",
        "NEUTRAL": "neutral",
        "POSITIVE": "positive"
    }

    sentiment = label_map[result[0]['label'].upper()]
    confidence = result[0]['score']
    return {
        "sentiment": sentiment,
        "confidence": confidence
    }

```

```python
text = "I love using this product! It's amazing."
print(predict_sentiment(text))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

def evaluate_model(true_labels, predicted_labels):
    # Calculate metrics
    accuracy = accuracy_score(true_labels, predicted_labels)
    precision = precision_score(true_labels, predicted_labels, average='weighted')
    recall = recall_score(true_labels, predicted_labels, average='weighted')
    f1 = f1_score(true_labels, predicted_labels, average='weighted')

    # Print the classification report
    report = classification_report(true_labels, predicted_labels)

    # Return the metrics
    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1_score": f1,
        "classification_report": report
    }

# Example usage
if __name__ == "__main__":
    # Example true labels and predicted labels
    true_labels = ["positive", "negative", "neutral", "positive", "negative"]
    predicted_labels = ["positive", "negative", "positive", "positive", "negative"]

    metrics = evaluate_model(true_labels, predicted_labels)
    print(metrics)
```
